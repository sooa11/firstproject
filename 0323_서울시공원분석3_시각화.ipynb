```python
import numpy as np
import pandas as pd
import json
import folium
```

```python
df = pd.read_csv('./자치구별 공원 현황.csv')
df.head()
```

```python
filename = './seoul-gu-simple.json'
with open(filename, encoding='utf-8') as json_file:
    gu_geo_data = json.load(json_file)
```

# 1. 자치구별 자치구면적 대비 공원면적
# 2. 자치구별 자치구인구 대비 공원면적
# 3. 공원위치 공원크기별로 원크기 표시

```python
# 자치구 면적대비 공원면적
center = [37.581, 126.986]
map = folium.Map(location=center, zoom_start=11,
                tiles='Stamen Toner')
folium.Choropleth( 
                   geo_data=gu_geo_data, data=df ,
                   columns=('지역', '공원면적'),
                   key_on='feature.id',
                   fill_color= 'BuPu',
                   legend_name='공원면적').add_to(map)
map
```

```python
park = pd.read_csv('서울공원요약.csv')
park.head()
```

```python
#자치구 인구대비 공원면적
center = [37.581, 126.986]
map = folium.Map(location=center, zoom_start=11,
                 tiles='Stamen Toner')
folium.Choropleth(
                    geo_data=gu_geo_data, data=df,
                    columns=('지역', '공원면적'),
                    key_on='feature.id',
                    fill_color='BuPu',
                    legend_name='공원면적').add_to(map)

for i in park.index:
    folium.CircleMarker([park.위도[i], park.경도[i]],
                       radius=int(park.크기[i]),
        tooltip=f'{park.공원명[i]}({park.면적[i]:,d})',
                       color='green', fill_color='green'
                       ).add_to(map)
map
```
